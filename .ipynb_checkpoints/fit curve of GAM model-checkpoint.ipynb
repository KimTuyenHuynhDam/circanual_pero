```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from pygam import LinearGAM, s
from sklearn.metrics import r2_score
import numpy as np
from scipy.stats import sem
import math
```

```python



def calculate_sem_excluding_nan(data):
    """
    Calculate the Standard Error of the Mean (SEM), excluding NaN values.

    Args:
    - data: NumPy array or pandas DataFrame with shape (n_samples, n_features).

    Returns:
    - sem_values: NumPy array containing the SEM values for each column.
    """
    if isinstance(data, pd.DataFrame):
        # Convert DataFrame to NumPy array for calculation
        data = data.values
    
    # Count the number of non-NaN values per column
    non_nan_counts = np.sum(~np.isnan(data), axis=0)
    
    # Calculate standard deviation excluding NaN, for each column
    std_devs = np.nanstd(data, axis=0, ddof=0)
    
    # Calculate SEM excluding NaN
    sem_values = std_devs / np.sqrt(non_nan_counts)
    
    return sem_values
```

```python
###GAM curve with option to select knots


# Set a seed value (use any integer value)
np.random.seed(100)  

def permutation_test(gam, X, y, num_permutations=1000):
    np.random.seed(100) 
    actual_r2 = r2_score(y, gam.predict(X))
    null_r2 = []
    for _ in range(num_permutations):
        y_permuted = np.random.permutation(y)
        gam.fit(X, y_permuted)
        null_r2.append(r2_score(y_permuted, gam.predict(X)))
    p_value = (np.sum(null_r2 >= actual_r2) + 1) / (num_permutations + 1)
    return p_value

def GAM_graph_knot_choice(data, num, num_knots=10, save_path=None):
    np.random.seed(100) 
    
    # Mapping numbers to month names
    month_names = {
        1: 'January', 2: 'February', 3: 'March', 4: 'April', 
        5: 'May', 6: 'June', 7: 'July', 8: 'August', 
        9: 'September', 10: 'October', 11: 'November', 12: 'December'
    }

    means = np.nanmean(data, axis=0)  # Calculate means of each column, excluding NaN

    means = np.where(np.isnan(means), 0, means)
    std_devs = calculate_sem_excluding_nan(data)  # Calculate SEM of each column, excluding NaN
    months = np.arange(1, 13)  # Months from 1 to 12


    plt.style.use('seaborn-v0_8-colorblind')
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.errorbar(months, means, yerr=std_devs, fmt='o', color='darkblue', ecolor='lightblue', elinewidth=2, capsize=5)

    # Create and fit a GAM
    gam = LinearGAM(s(0, n_splines=num_knots)).fit(months, means)

    # Predict using the GAM for smooth curve
    XX = np.linspace(1, 12, 100)
    smooth_means = gam.predict(XX.reshape(-1, 1))  # Reshape XX for prediction

    # Plot the smooth curve
    ax.plot(XX, smooth_means, label='_Hidden Label') # label=f'Smooth Curve Fit (Knots={num_knots})'

    # Calculate R-squared (Fitness Score)
    predictions = gam.predict(months)
    fitness_score = r2_score(means, predictions)

    # Estimate p-value using permutation test
    p_value = permutation_test(gam, months, means)

    # Calculate the proposed upper limit for the y-axis
    proposed_upper_ylim = max(means) + max(std_devs) * 1.5

    # Set the minimum upper limit value
    min_upper_ylim = 20

    # Determine the actual upper limit by comparing the proposed limit with the minimum limit
    actual_upper_ylim = max(proposed_upper_ylim, min_upper_ylim)

    # Round actual_upper_ylim up to the nearest multiple of 5
    actual_upper_ylim = math.ceil(actual_upper_ylim / 5) * 5

    # Display Fitness Score and p-value on the graph
    # Adjust the placement of fitness score and p-value text to be at the bottom left inside the plot
    ax.text(0.05, 0.95, f'Fitness Score (R-squared): {fitness_score:.3f}\nP-value: {p_value:.3f}',
    horizontalalignment='left', verticalalignment='top', transform=ax.transAxes, 
     fontsize=10, bbox=dict(facecolor='white', alpha=0.8))
    
    # Display Fitness Score and p-value in the plot margin
    #text_str = f'Fitness Score (R-squared): {fitness_score:.3f}\nP-value: {p_value:.3f}'
    #plt.gcf().text(0.02, 0.95, text_str, fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
    
    # Set the y-axis ticks to multiples of 5
    ax.yaxis.set_major_locator(MultipleLocator(5))

    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.set_xlabel('Litter in', fontsize=12)
    ax.set_ylabel('% offspring in month',fontsize=12)
    ax.set_title(month_names[num], fontsize=14)
    ax.set_xticks(months)
    #ax.set_xticklabels(list(month_names.values()), rotation=45)  # Optional: To show month names on the x-axis
    ax.set_xlim(0.5, 12.5)

    ax.set_ylim(0, actual_upper_ylim)
    ax.legend()  # Add a legend to explain the smooth curve

    if save_path:  # If 'save_path' is provided after displaying the plot, save the plot
        plt.savefig(save_path)  # Save the plot with the specified path and format

    plt.show()  # Show the plot after saving (if save_path is provided)


```

```python

np.random.seed(100) 
# Function to compute GAM scores and return R-squared and p-value

def GAM_graph_knot_choice_score(data, num_knots=10):
    np.random.seed(100) 
    
    means = np.nanmean(data, axis=0)  # Calculate means of each column, excluding NaN
    means = np.where(np.isnan(means), 0, means)
    
    months = np.arange(1, 13)  # Months from 1 to 12

    # Create and fit a GAM
    gam = LinearGAM(s(0, n_splines=num_knots)).fit(months, means)
    
    # Calculate R-squared (Fitness Score)
    predictions = gam.predict(months)
    fitness_score = r2_score(means, predictions)

    # Estimate p-value using permutation test
    p_value = permutation_test(gam, months, means)

    return fitness_score, p_value

# Function to process Dam/Sire data for a species
def process_data(data, species_name, data_type='Dam'):
    np.random.seed(100) 
    all_scores = []
    month_scores = []
    
    # Assuming there's only one column that includes "BirthMonth" in its name
    birth_month_columns = df.filter(like='BirthMonth').columns
    
    parent_column = birth_month_columns[0]
    
    for month in range(1, 13):
        values=[]
        # Filter data for the current month
        values = data[data[parent_column] == month].iloc[:, 1:]  # Exclude 'BirthMonth'  
         # Remove rows where all elements are NaN
        values = values.dropna(how='all')
    # produce graph
        filename = f'./GAM/12 knots/{species_name} -{data_type} born in {month} (SMF 12 knots).png'  # Unique filename based on the count

        GAM_graph_knot_choice(values, month, num_knots=12, save_path=filename) 
         
            
    # Get scores for each month
        fitness_score, p_value = GAM_graph_knot_choice_score(values, num_knots=12)
        month_scores.append({'Month': month, 'Fitness Score': fitness_score, 'P-value': p_value})
        
    
        
    # Store scores for all months in a list
    all_scores.extend(month_scores)
        
    # Create a DataFrame for scores of Dams/Sires of a species
    scores_df = pd.DataFrame(all_scores)
    scores_df['Species'] = species_name  # Add species_name as a column
    #scores_df['Dam/Sire'] = data_type  # Add species_name as a column
    return scores_df

```
